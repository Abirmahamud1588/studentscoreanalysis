```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("std.csv")
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
plt.figure(figsize=(8, 6))
ax= sns.countplot(x='Gender', data=df )
ax.bar_label(ax.containers[0])
plt.title('Number of Males and Females in the Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
average_scores = df.groupby('Gender').agg({'MathScore': 'mean'})
print(average_scores)
```

```python
plt.figure(figsize=(5, 5))
ax= sns.barplot(x='Gender',y="MathScore", data=df )
ax.bar_label(ax.containers[0])
plt.title("Average Score for Math Genderwise")
plt.ylabel('MATH Score')
plt.xlabel('GENDER')
plt.show()
```

Male got the more average scores in math compared to Female

```python
plt.figure(figsize=(5, 5))
ax= sns.barplot(x='Gender',y="ReadingScore", data=df )
ax.bar_label(ax.containers[0])
plt.title("Average Score for Reading Genderwise")
plt.ylabel('Reading Score')
plt.xlabel('GENDER')
plt.show()
```

 Female got the more average scores in Reading compared to male

```python
plt.figure(figsize=(5, 5))
ax= sns.barplot(x='Gender',y="WritingScore", data=df )
ax.bar_label(ax.containers[0])
plt.title("Average Score for writting Genderwise")
plt.ylabel('Writting Score')
plt.xlabel('GENDER')
plt.show()
```

 Female got the more average scores in writting compared to male

```python
gp=df.groupby('ParentEduc').agg({'MathScore': 'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.show()
```

parents who got master's degree ,have good impact on performance of their children

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels=["group A", "group B", "group C", "group D", "group E"], autopct='%1.1f%%')
plt.title("Ethnic Group Distribution")
plt.show()
```
